```python
#remove tags 
import re 
def remove_tags(raw_text):
    clean_text = re.sub(re.compile('<?.*>'),' ',raw_text)
    return clean_text
```

```python
imdb_df['review'] = imdb_df['review'].apply(remove_tags)
```

```python
imdb_df.head()
```

```python
# stopword removal 
from nltk.corpus import stopwords 
```

```python
stop_words = stopwords.words('english')
```

```python
imdb_df['review'] = imdb_df['review'].apply(lambda x:[word for word in x.split() if word not in stop_words]).apply(lambda x:' '.join(x))
```

```python
from gensim.utils import simple_preprocess  
```

```python
from nltk import sent_tokenize
```

```python
story = []
for doc in imdb_df['review']:
    gen_text = simple_preprocess(doc)
    story.append(gen_text)
```

```python
story
```

```python
import gensim
```

```python
model = gensim.models.Word2Vec(window=10,min_count=2)
```

```python
model.build_vocab(story)
```

```python
model.train(story, total_examples=model.corpus_count, epochs=model.epochs)
```
